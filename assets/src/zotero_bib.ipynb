```python
"""Liquid syntax generator for representing Zotero libraries on github pages."""
# Using python to generate liquid to generate html: syntaxception
# Unnecessarily complex; Mostly for practice
# Export zotero library: File -> Export Library -> Format: CSL JSON

BIB_PATH = 'zotero.json'
EXPORT_PATH = 'biblio.html'

import json
from contextlib import contextmanager

vars = sorted(vars)
types = sorted(types)

@contextmanager
def _for():
    print('{% for b in site.data.zotero %}')
    yield
    print('{% endfor %}')

    
@contextmanager
def _case(c):
    print('{% case "' + c + '" %}')
    yield
    print('{% endcase %}')

_title = lambda t: print("<h2 class='title'>{{" + 't' + "}}</h2>")
_when = lambda w: print('{% when "' + w + '" %}')
_assign = lambda v: print('{% assign ' + v + ' = b.' + v + ' %}')
_table_row = lambda x: print("<tr><th class='" + x.lower().replace(' ','-') + "'>" + x.capitalize() + "</th><td>{{" + x + "}}</td></tr>")
_li = lambda: print('<li id="item_{{type}}" class="">')
with open(BIB_PATH, 'r') as f:
    bib = json.load(f)
    types = {}
for b in bib:
    if b['type'] not in types:
        var = set(v for v in b.keys())
        types[b['type']] = var
```

```python
if __name__ == '__main__':    
    print('<ul>')
    with _for():
        with _case('type'):
            for t,v in types.items():
                _when(t)
                _li()
                print('<table><tbody>')
                _table_row(t)
#                for v in vars:
#                    {% for a in author %}
#            <tr><th class='author'>Author</th><td>{{a}}</td></tr>
#          {% endfor %}
                print('</tbody></table>')
                print('</li>')

    
    print('</ul>')
```
